### Are plot are very good when we are tracking the changing between 2 or 3 groups

```python
import matplotlib.pyplot as plt
```

```python
days=[1,2,3,4,5]
```

```python
sleeping=[7,8,6,11,7]
```

```python
eating=[2,3,4,3,2]
```

```python
working=[7,8,7,2,2]
```

```python
playing=[8,5,7,8,12]
```

```python
plt.plot([],[],color='m',label='Sleeping',linewidth=5)
plt.plot([],[],color='c',label='Eating',linewidth=5)
plt.plot([],[],color='r',label='Working',linewidth=5)
plt.plot([],[],color='k',label='Playing',linewidth=5)

plt.stackplot(days,sleeping,eating,working,eating,working,playing,colors=['m','c','r','k'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('stck Plot')
plt.legend()
plt.show()
```
