```python
from matplotlib import pyplot as plt
```

```python
from matplotlib import style
```

```python
style.use('ggplot')
```

```python
x=[5,8,10]
```

```python
y=[12,16,6]
```

```python
x2=[9,6,11]
```

```python
y2=[6,15,7]
```

```python
plt.plot(x,y,'g',label='Line One',linewidth=5)
plt.plot(x2,y2,'c',label='Line Two',linewidth=2)
plt.title("Epic Enfo")
plt.ylabel("y axis")
plt.xlabel("x axis")
plt.legend()
plt.grid(True,color='w')
plt.show()
```
