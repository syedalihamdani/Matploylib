### subplot()

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# Compute the x and y coordinates for points on sine and cosine curves
```

```python
x=np.arange(0,3*np.pi,0.1)
```

```python
print(x)
```

```python
y_sin=np.sin(x)
```

```python
y_cos=np.cos(x)
```

```python
# Set up a subplot grid that has height 2 and width 1,
```

```python
# and set the first such subplot as active.
```

```python
plt.subplot(2,1,1)
```

```python

# Make the first plot
```

```python
plt.plot(x,y_sin)
```

```python

plt.title('Sine')
```

```python

# Set the second subplot as active,and make the second plot
```

```python
plt.subplot(2,1,2)
```

```python
plt.plot(x,y_cos)
```

```python
plt.title('Cosine')
```

```python
plt.show()
```
