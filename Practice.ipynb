```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
slices=[33,49,22]
```

```python
Names=["new","Old","Moderate"]
```

```python
col=['b','r','g']
```

```python
plt.pie(slices,
       labels=Names,
       colors=col,
       startangle=90,
       shadow=True,
       explode=(0,0.1,0))
plt.title("pie plot")
plt.show()
```

```python
plt.pie(slices,
       labels=Names,
       colors=col,
       startangle=90,
       shadow=True,
       explode=(0.1,0,0))
plt.title("Pie Chart")
plt.show()
```

```python
plt.pie(slices,
       labels=Names,
       colors=col,
       startangle=90,
       shadow=True,
       explode=(0,0,0.1))
plt.title("Pie Chart")
plt.show()
```
