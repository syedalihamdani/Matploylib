### Quiver show the velocity points in the form of arrows

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x,y =np.meshgrid(np.arange(-2,2,.2), np.arange(-2,2,.25))
```

```python
z=x*np.exp(-x**2 -y**2)
```

```python
v,u =np.gradient(z,.2,.2)
```

```python
fig,ax=plt.subplots()
q=ax.quiver(x,y,u,v)
plt.show()
```
